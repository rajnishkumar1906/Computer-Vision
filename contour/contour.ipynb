```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("../Images/shapes.png")

image_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

image_gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(image_rgb)
```

```python
plt.imshow(image_gray,cmap='gray')
```

```python
gray = cv2.Canny(image_gray,80,150)
plt.imshow(gray,cmap='grey')
```

```python
contours , hierarchy = cv2.findContours(gray,cv2.RETR_LIST,cv2.CHAIN_APPROX_SIMPLE)

# RETR_LIST - mode  -  contour retroval mode
# Mode determines hierarchial relationship between contours
# methos - decides how contour points are stored.

contour_image = image.copy()
cv2.drawContours(contour_image,contours,-1,(0,255,0),5)
                                            #color  #thikness
# -1 is value of index parameter . for no specifix index we have used this value


plt.figure(figsize=(10,5))

plt.subplot(1,3,1)
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.title("Original image")

plt.subplot(1,3,2)
plt.imshow(gray,cmap='gray')
plt.title("Mask")

plt.subplot(1,3,3)
plt.imshow(cv2.cvtColor(contour_image,cv2.COLOR_BGR2RGB))
plt.title("Contour image")

plt.show()
```

```python
# Any pixel value greater than 200 will  be treated as 255  
retrval , binary  = cv2.threshold(image_gray,200,255,cv2.THRESH_BINARY)  

contours , hierarchy = cv2.findContours(binary,cv2.RETR_LIST,cv2.CHAIN_APPROX_SIMPLE)

# RETR_LIST - mode  -  contour retroval mode
# Mode determines hierarchial relationship between contours
# methos - decides how contour points are stored.

contour_image = image.copy()
cv2.drawContours(contour_image,contours,-1,(0,255,0),5)
                                            #color  #thikness
# -1 is value of index parameter . for no specifix index we have used this value

plt.figure(figsize=(2,3))
plt.imshow(binary,cmap='gray')
plt.title("binary image")

plt.figure(figsize=(10,5))

plt.subplot(1,3,1)
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.title("Original image")

plt.subplot(1,3,2)
plt.imshow(binary,cmap='gray')
plt.title("Binary Image")

plt.subplot(1,3,3)
plt.imshow(cv2.cvtColor(contour_image,cv2.COLOR_BGR2RGB))
plt.title("Contour image")

plt.show()
```

```python
import numpy as np

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Convert to HSV (for color detection)
image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Define HSV color range
# Targetting light blue, light purple and dark purple colors
lower_bound = np.array([90, 50, 50], dtype=np.uint8)
upper_bound = np.array([150,205,255], dtype=np.uint8)

# Create binary mask where color range is met
mask = cv2.inRange(image_hsv, lower_bound, upper_bound)
plt.imshow(mask,cmap='grey')

# Find contours in mask
contours, hierarchy = cv2.findContours(mask, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)

# Draw contours on a copy of the image
contour_image = image.copy()
cv2.drawContours(contour_image, contours, -1, (0, 255, 0), -1)  # green contours

# Display
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(binary, cmap='gray')
plt.title("Binary Mask (Thresholded Color)")
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(cv2.cvtColor(contour_image, cv2.COLOR_BGR2RGB))
plt.title("Contour Image")
plt.axis('off')

plt.tight_layout()
plt.show()
```
