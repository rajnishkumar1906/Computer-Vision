```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("../READING_AND_WRITING_IMAGES/mountain.png")
image_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
print(image_rgb.shape)
```

```python
channels = cv2.split(image_rgb)
colors = ['red','green','blue']
for channels , color in zip(channels,colors):
    hist = cv2.calcHist([channels],[0],None,[256],[0,256])
    plt.plot(hist,color=color)
plt.title("Color Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.show()
```

```python
# Convert from BGR to HSV
image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Split into channels
h, s, v = cv2.split(image_hsv)

# Calculate histograms for each channel
h_hist = cv2.calcHist([h], [0], None, [180], [0,180])
s_hist = cv2.calcHist([s], [0], None, [256], [0,256])
v_hist = cv2.calcHist([v], [0], None, [256], [0,256])

# Plot
fig, axes = plt.subplots(1, 3, figsize=(15,5))

axes[0].plot(h_hist, color='orange')
axes[0].set_title("Hue Histogram")
axes[0].set_xlabel("Hue value")
axes[0].set_ylabel("Frequency")

axes[1].plot(s_hist, color='green')
axes[1].set_title("Saturation Histogram")
axes[1].set_xlabel("Saturation value")
axes[1].set_ylabel("Frequency")

axes[2].plot(v_hist, color='blue')
axes[2].set_title("Value (Brightness) Histogram")
axes[2].set_xlabel("Value intensity")
axes[2].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```
