```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("../Images/rajnishk.png")
# image = cv2.imread("../Images/bholenath.jpg")
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray,cmap='gray')
```

### Average blurring

```python
kernel_s = 10

average_blur = cv2.blur(image_gray,(kernel_s,kernel_s))

plt.imshow(average_blur,cmap='gray')
plt.title("Average blur")
plt.show()
```

### Gaussian Blurring

```python
# for gaussina blurring , kernel is or odd size
kernel_s = 31
gaussian_blur = cv2.GaussianBlur(image_gray,(kernel_s,kernel_s),0)
plt.imshow(gaussian_blur,cmap='grey')
plt.title("Gaussian blur")
plt.show()
# grey and gray are both same
```

### Median Blurring

```python
median_blur = cv2.medianBlur(image_gray,3)
plt.imshow(median_blur,cmap='gray')
plt.title("Median blur")
plt.show()
```

```python
plt.figure(figsize=(14,12))

plt.subplot(2,2,1)
plt.title("original Image")
plt.imshow(image_gray,cmap='gray')
plt.axis('off')


plt.subplot(2,2,2)
plt.title("Average blur Image")
plt.imshow(average_blur,cmap='gray')
plt.axis('off')


plt.subplot(2,2,3)
plt.title("Gaussian Blur Image")
plt.imshow(gaussian_blur,cmap='gray')
plt.axis('off')


plt.subplot(2,2,4)
plt.title("Median Blur Image")
plt.imshow(median_blur,cmap='gray')
plt.axis('off')

plt.show()
```
