```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("../Images/rajnishk.png", cv2.IMREAD_GRAYSCALE)

# Sobel edge detector
sobel_x = cv2.Sobel(image,cv2.CV_64F,1,0,ksize=3)
sobel_y = cv2.Sobel(image,cv2.CV_64F,0,1,ksize=3)

sobel_combined = cv2.magnitude(sobel_x,sobel_y)

# Laplacian edge detector 
laplacian = cv2.Laplacian(image, cv2.CV_64F, ksize=3)
laplacian_abs = cv2.convertScaleAbs(laplacian)

# canny edge detector 
canny_edges = cv2.Canny(image, 100, 200)
```

```python
laplacian_abs
```

```python
titles = ["Original Image","Sobel Image","Laplacian all","Laplacian Absolute","Canny"]
images = [image,sobel_combined,laplacian,laplacian_abs,canny_edges]

plt.figure(figsize=(12,8))
for i in range(len(titles)):
    plt.subplot(2,3,i+1)
    if titles[i] == "Laplacian all":
        plt.imshow(images[i], cmap='gray', vmin=-255, vmax=255) 
    else:
        plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')
plt.tight_layout()
plt.show()
```
