```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("./giraffe-kenya.png")
```

```python
# cv2 will show image in rgb but here plt will be showing in bgr as it its current format
plt.imshow(image)
print("image will be in bgr as open cv loads image in bgr")
```

```python
image_rbg = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
image_hsv = cv2.cvtColor(image,cv2.COLOR_BGR2HSV)
```

```python
plt.imshow(image_rbg)
```

```python
plt.imshow(image_hsv)
```

```python
# hue :saturation ,value
lower_black= np.array([0,0,0]) 
upper_black = np.array([100,80,255])
```

```python
mask = cv2.inRange(image_hsv,lower_black,upper_black)
```

```python
plt.imshow(mask)
```

```python
image_result = image.copy()

# Replace black areas (where mask>0) with white (255,255,255 in BGR)
image_result[mask>0] = np.array([255,255,255])
```

```python
# Convert images from BGR to RGB for displaying with matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_result_rgb = cv2.cvtColor(image_result, cv2.COLOR_BGR2RGB)
```

```python
# Show original and result side by side
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(image_rgb)
plt.axis('off')
```

```python
plt.subplot(1,2,2)
plt.title('Black replaced with White')
plt.imshow(image_result_rgb)
plt.axis('off')

plt.show()
```
