```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("./bholenath.png")
image_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
print(image_rgb.shape)
```

### Cropping

```python
x_start, y_start = 400, 450
x_end, y_end = 1000, 736
shivling = image_rgb[y_start:y_end, x_start:x_end]

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image_rgb, cv2.COLOR_BGR2RGB))
plt.title("Bholenath (Original)")
plt.axis('off')

cv2.imwrite("../Images/bholenath.jpg", image_rgb)

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(shivling, cv2.COLOR_BGR2RGB))
plt.title("Shivling (Cropped)")
plt.axis('off')

plt.tight_layout()
plt.show()
```

### Flipping

```python
# 1  --> horizontal
# 0  --> vertically
# -1 --> both axes
flip_horizontal = cv2.flip(image,1)
flip_vertical = cv2.flip(image,0)
flip_both = cv2.flip(image,-1)

titles = ["Original image","Horizontal flip","Vertical Flip","Flip both sides"]

images = [
    image_rgb,
    cv2.cvtColor(flip_horizontal,cv2.COLOR_BGR2RGB),
    cv2.cvtColor(flip_vertical,cv2.COLOR_BGR2RGB),
    cv2.cvtColor(flip_both,cv2.COLOR_BGR2RGB)
    ]


plt.figure(figsize=(12,8))
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')
plt.tight_layout()
plt.show()
```

### Rotation

```python
rotate_90_clockwise = cv2.rotate(image_rgb,cv2.ROTATE_90_CLOCKWISE)
rotate_90_counterclockwise = cv2.rotate(image_rgb,cv2.ROTATE_90_COUNTERCLOCKWISE)
rotate_180 = cv2.rotate(image_rgb,cv2.ROTATE_180)


titles = ["Original Image","90 degree clockwise","90 degree counter clockwise","180 degree rotation"]
images = [
    image_rgb,
    rotate_90_clockwise,
    rotate_90_counterclockwise,
    rotate_180
    ]


plt.figure(figsize=(12,8))
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')
plt.tight_layout()
plt.show()
```
