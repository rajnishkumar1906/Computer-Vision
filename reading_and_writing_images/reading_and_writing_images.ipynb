```python
import cv2
```

```python
# Read thee image
image  = cv2.imread("./mountain.png")
```

```python
if image is None:
    print("Error : could not load he image")
else:
    print("Image loaded successfully")
```

```python
# Display the image
cv2.imshow("Image window",image)
cv2.waitKey(0)    #pause the program uintil any key is clicked
cv2.destroyAllWindows()    #Destroying the image 
```

```python
# Write the image

image_new = image + 100    #image is a numpy array
cv2.imwrite("./output.png",image_new)
```
