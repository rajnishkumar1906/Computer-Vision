```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
height , width = 500,1000
skyblue = (255,177,0)   # bgr format naturally
canvas = np.full((height,width,3),skyblue,dtype=np.uint8)
# color change
canvas = cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB)
plt.imshow(canvas)
```

```python
text = "Hello Computer vision"
font = cv2.FONT_HERSHEY_COMPLEX
# coordinate for bottom left corner of the text
origin = (200,250)
font_scale = 1.5
color = (255,255,255)
thickness = 2
cv2.putText(canvas,text,origin,font,font_scale,color,thickness,cv2.LINE_AA)

plt.imshow(canvas)
```
