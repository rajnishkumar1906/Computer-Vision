```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# create an emoty canvas
height , width = 500 , 700
blue = (255,127,0)    #rgb but it will be read as bgr
canvas = np.full((height,width,3),blue,dtype=np.uint8)
# convert bgr to rgb
plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
```

### Draw a line

```python
# frame ,  start point , end point , color ,thickness
cv2.line(canvas,(20,20),(600,600),(255,255,255),3)
plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
```

### Draw a rectangle

```python
# last param is thickness (if -1 then complete filling)
cv2.rectangle(canvas,(100,100),(300,300),(200,200,200),-1)
plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
```

### Draw a circle

```python
# frame , centre , radius , color , thickness
cv2.circle(canvas,(400,400), 100,(0,0,255),2)
plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
```

### Draw an ellipse

```python
# frame, center , major and minor axis
cv2.ellipse(canvas,(350,350),(100,50),45,0,360,(0,255,0),3)
plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
```

### Draw a polygon

```python
points = np.array([[400,0],[200,100],[600,200]],np.int32)
new_reshaped_point = points.reshape((-1,1,2))

print(new_reshaped_point)

# isclosed can be true or false (if its false then there will be linkage between points bit open figure)
cv2.polylines(canvas, [new_reshaped_point], isClosed=True, color=(0,255,0), thickness=3)
plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
```
