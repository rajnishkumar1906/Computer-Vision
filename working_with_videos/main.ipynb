```python
import cv2
```

```python
# Load the video
video = cv2.VideoCapture("./mountain.mp4")
```

```python
# Display the video
while True:
    # ret captures boolen value
    # frame is in numpy
    ret , frame = video.read()
    if not ret: 
        break
    cv2.imshow("video frame",frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

video.release()
cv2.destroyAllWindows()
    
```

```python
# Write a video
# * representing codec
# codec is used to compress or decrompress a video
fourcc = cv2.VideoWriter_fourcc(*"mp4v")

width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))
out = cv2.VideoWriter("./output.mp4",fourcc,30,(width,height))

while True:
    ret , frame = video.read()
    
    if not ret:
        break
    ''' we will increamen frame after checkign ret so that it becomes 
    confirm whether image exists'''
    frame += 20
    out.write(frame)
    
out.release()
video.release()
cv2.destroyAllWindows()
```
